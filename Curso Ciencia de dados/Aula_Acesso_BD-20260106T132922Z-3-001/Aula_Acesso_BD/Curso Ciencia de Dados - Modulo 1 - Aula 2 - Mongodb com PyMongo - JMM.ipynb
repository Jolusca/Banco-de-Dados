### Junção de Tabelas

```python
# Iremos colocar a tabela TIPO_ESTABELECIMENTO dentro do DataFrame
df = pd.DataFrame.from_records(tipo_estabelecimento.find())
df3
```

```python
df3.shape
```

```python
# Faremos o JOIN das duas tabelas através da coluna "RES_TIP_ESTABELECIMENTO" e "_ID" e armazenaremos o resultado em um novo DF
df4 = pd.merge(df, df3, left_on = "RES_TIP_ESTABELECIMENTO", right_on="RES_TIP_ESTABELECIMENTO")
```

```python
# Vamos inserir uma nova instância na tabela ESTABELECIMENTO
NovoEstabelecimento = {"RES_ID" : 4, 
"RES_NOM_ESTABELECIMENTO" : "Ânimo Cafe", 
"RES_ENDERECO" : "Av. Pontes Vieira",
"RES_NUMERO" : "417",
"RES_COMPLEMENTO" : "",
"RES_CEP" : "60013523",
"RES_BAIRRO" : "São João do Tauape",
"RES_CIDADE" : "Fortaleza",
"RES_UF" : "CE",
"RES_TELEFONE" : "",
"RES_TIP_ESTABELECIMENTO" : 3
}
```

```python
estabelecimento.insert_one(NovoEstabelecimento)
```

```python
# Verificando inserção
df = pd.DataFrame.from_records(estabelecimento.find())
df
```

```python
# Removendo instância
estabelecimento.delete_one({"RES_NOM_ESTABELECIMENTO": "Ânimo Cafe"})
```

```python
# Verificando remoção
df = pd.DataFrame.from_records(estabelecimento.find())
df
```
